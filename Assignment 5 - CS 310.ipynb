## Part 1: Visualize Daily Treasury Yield Curve Rates

### (10pts) 1. You are provided with a Daily Treasury Yield Curve Rates as rates.csv file. Load into dataframe. Use Date as index.

```python
# Stefana Ciustea - CS 310 - Assignment 5
# 11/20/2022

# your code here:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

rates = pd.ExcelFile('rates.xlsx')
df_rates = rates.parse('rates')
df_rates = df_rates.set_index('Date')
```

```python
print(df_rates)
```

### (10pts) 2. Plot curves on all columns in one graph, with x axis the count of date, y axis the rates. By visually inspecting the curves, which curve has the overall highest rate through out the year? Highlight that curve with red solid line, every other curves in dash line.

```python
# your code here:
xi = list(range(len(df_rates)))
plt.plot(xi, df_rates['1 Mo'], label = "1 Mo", linestyle = 'dashed')
plt.plot(xi, df_rates['2 Mo'], label = "2 Mo", linestyle = 'dashed')
plt.plot(xi, df_rates['3 Mo'], label = "3 Mo", linestyle = 'dashed')
plt.plot(xi, df_rates['6 Mo'], label = "6 Mo", linestyle = 'dashed')
plt.plot(xi, df_rates['1 Yr'], label = "1 Yr", linestyle = 'dashed')
plt.plot(xi, df_rates['2 Yr'], label = "2 Yr", linestyle = 'dashed')
plt.plot(xi, df_rates['3 Yr'], label = "3 Yr", linestyle = 'dashed')
plt.plot(xi, df_rates['5 Yr'], label = "5 Yr", linestyle = 'dashed')
plt.plot(xi, df_rates['7 Yr'], label = "7 Yr", linestyle = 'dashed')
plt.plot(xi, df_rates['10 Yr'], label = "10 Yr", linestyle = 'dashed')
plt.plot(xi, df_rates['20 Yr'], label = "20 Yr", linestyle = 'dashed')
plt.plot(xi, df_rates['30 Yr'], label = "30 Yr", linestyle = 'dashed')
plt.plot(xi, df_rates['30 Yr'], label = "30 Yr", linestyle = 'solid', color = 'r')
plt.legend() 
plt.show()
```

### (10pts)3. PairGrid all short rates (less than one year rate). You should get a 4x4 graph. Use histplot for diagonal plots, use scatter plot for upper corner, and regplot for lower corner.

```python
# your code here:
import seaborn as sns
%matplotlib inline

d = sns.pairplot(df_rates, vars = ['1 Mo', '2 Mo', '3 Mo', '6 Mo'])
d.map_lower(plt.scatter, alpha = 0.6)
```

### (10pts)4. "reg" jointplot 2m and 3m rates.

```python
# your code here
t = sns.jointplot(data=df_rates, x='2 Mo', y='3 Mo', kind="reg")
```

### (10pts)5. Use heatmap to show the correlations between each column.

```python
# your code here:
corr = df_rates.corr()
heat_map = sns.heatmap(corr)
```
